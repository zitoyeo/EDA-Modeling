## 3. 데이터 정비하기
### 3-1 시간 타입의 컬럼 정비하기

```python
#date 컬럼은 int64, time컬럼은 object 타입으로 date컬럼의 타입을 변환한 후 두 컬럼을 합친다.

lakers['date'] = lakers['date'].apply(str)
lakers['datetime'] = lakers['date'] + ' ' + lakers['time']

lakers.head()
```

```python
# date컬럼과 새로 만든 datetime 컬럼을 날짜 타입으로 바꿔준다
# 날짜 타입은 연, 월, 일, 시, 분,초 등의 정보가 보기 편하다(시계열 분석시 편리)
lakers['datetime'] = pd.to_datetime(lakers['datetime'])
lakers['date'] = pd.to_datetime(lakers['date'], format="%Y%m%d")
lakers.head()
```

# 경기결과 집계 하고 시계열 그래프로 그리기

```python
# 날짜별 경기 결과를 group by로 계산해 보자
# 결과 컬럼에서 성공했을때의 득점만 더하면 경기의 총 점수ㅏㄱ 나온다.

made = lakers['result'].isin(['made'])
lakers[made].head()
```

```python
# 날짜별 팀별 슛 성공 건수를 집계해보면

df3 = lakers[made][['date', 'team', 'points']].groupby(['date', 'team'], as_index=True).sum()
df3
```

```python
#df3.sort_values(by='points', ascending=False)
df3.reset_index(inplace =True)
```

```python
# la 레잌스와 상대편으로 경기 결과 나누기
lal = df3[df3['team'] == 'LAL']
oop = df3[df3['team'] != 'LAL']
```

```python
lal
```

```python
oop
```

```python
# 경기 겨과를 가로 방향으로 붙이기
# pd.merge 함수 사용  : 불일때 key는 ? = 날짜 (어떤 기준으로 붙일지 기준)

df4 = pd.merge(lal, oop, on='date')
df4 = df4.rename(columns={'team_x' : 'LAL', 'team_y' : 'Opp', 'points_x' : 'LAL_Points', 'points_y' : 'Opp_points'})
df4
```

```python
df4.shape[0]
```

```python
df4['Opp_points'].shape[0]
```

```python
# la 레이커스가 이긴날과 진날 고르기
win = df4[df4['LAL_Points'] > df4['Opp_points']]
lose = df4[df4['LAL_Points'] < df4['Opp_points']]

print("LA 레이커스는 총 {} 회의 경기중, {}번 경기에서 이겼다".format(df4.shape[0], df4[df4['LAL_Points'] > df4['Opp_points']].shape[0]))
```

```python
# 경기 결과를 시계열 그래프로 확인하기
plt.figure(figsize=(18,5))
plt.title("08 - 09 Season : LA LAkers", fontsize=15)

#la 레이커스 점수를 오렌지 실선으로 나타낸다

plt.plot(lal['date'], lal['points'], '-', alpha=1, color = 'orange', label='LA Lakers')

# la 레이커스가 이긴날은 빨간점, 진날은 파란점으로 표시
plt.plot(win['date'], win['LAL_Points'], 'o', alpha=.6, color='red')
plt.plot(lose['date'], lose['LAL_Points'], 'o', alpha=.6, color='blue')

# 상대팀 점수를 초록색 점선으로 나타낸디ㅏ
plt.plot(oop['date'], oop['points'], '--' , alpha=.6, color='green', label='Opposite team')
plt.legend(loc='upper left', borderaxespad=0, fontsize=13)
plt.xticks(rotation=45)

plt.show()
```

# la 레이커스 팀의 코트 위치별 행동

```python
# LA lakers의 데이터만 가져오키 team이 la인거

lal2 = lakers[lakers['team'] == 'LAL']
lal2
```

```python
# 슛한 위치 확인하기
# 득점은 어느 위치에서 발생?
sns.set_style('whitegrid')
f, ax = plt.subplots(1,2, figsize=(12, 7), dpi =80)

#득점 성공 / 실패 분포
g1 = sns.scatterplot(data=lal2, x='x', y='y', hue='result', alpha=0.3, palette='Paired', ax=ax[0])
g2 = sns.kdeplot(data=lal2, x='x', y='y', hue='result', palette='Paired', ax=ax[0])

#득점 점수 별 분포
g3 = sns.scatterplot(data=lal2, x='x', y='y', hue='points', alpha=0.3, palette='Set3', ax=ax[1])
g4 = sns.kdeplot(data=lal2, x='x', y='y', hue='points', palette='Set3', ax=ax[1])

plt.show()
```

```python
# 세부 행동별 슛위치는 어떻게 다를까?
# 가장 많이 한 세부 행동 10개 고르기

lal3= lal2[lal2['type'].isin(lal2['type'].value_counts().head(10).index)]
lal3
```

```python
# 세부 행동별 위치 확인하기
sns.set_style('whitegrid')

f, ax = plt.subplots(1,1, figsize=(5,7), dpi = 80)

sns.scatterplot(data=lal3, x='x', y='y', hue=lal3['type'], palette='Set3', ax=ax)

#Legend가 10개기때문에 그래프안에 넣기 너무 많아 옆으로 뺴주자
ax.legend(title='Type', bbox_to_anchor=(1.03,1))

plt.show()
```
